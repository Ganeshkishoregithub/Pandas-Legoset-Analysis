```python
import pandas as pd
```

```python
#What percentage of all licensed sets ever released were starwars  themed?
```

```python
lego_sets=pd.read_csv('./datasets/lego_sets.csv')
lego_sets.head(10)
```

```python
themes=pd.read_csv('./datasets/parent_themes.csv')
themes.head(10)
```

```python
merged=lego_sets.merge(themes,left_on='parent_theme',right_on='name')
merged.drop(columns='name_y',inplace=True)
#merged=merged.dropna'set_num'
```

```python
#To find licensed 
licensed=merged[merged['is_licensed']==True]
licensed
```

```python
#They have mentioned as some set_num has zeros and we need to drop them alone
licensed=licensed.dropna(subset='set_num')
licensed
```

```python
starwars=licensed[licensed['parent_theme']=='Star Wars']
starwars
```

```python
starwars.shape[0]
```

```python
licensed.shape[0]
```

```python
#Percentage of starwars in licensed
the_force=int((starwars.shape[0]/licensed.shape[0])*100)
the_force
```

```python
licensed.head(10)
```

```python
licensed_sorted=licensed.sort_values('year')
licensed_sorted['count']=1
licensed_sorted
summed_df=licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head(50)
max_df=summed_df.sort_values('count',ascending=False,).drop_duplicates(['year'])
max_df.sort_values('year',inplace=True)
max_df
```
